```python
# Clean the dataset

divvy_cleaned = divvy_df.dropna()
```

```python
divvy_cleaned.shape
```

```python
divvy_cleaned.isnull().sum()
```

```python
divvy_cleaned.loc[:, 'started_at'] = pd.to_datetime(divvy_cleaned['started_at'])
divvy_cleaned.loc[:, 'ended_at'] = pd.to_datetime(divvy_cleaned['ended_at'])
```

```python
divvy_cleaned.dtypes
```

```python
divvy_cleaned.loc[:, 'ride_length'] = divvy_cleaned['ended_at'] - divvy_cleaned['started_at']
```

```python
divvy_cleaned.head(3)
```

```python
divvy_cleaned.dtypes
```

```python
divvy_cleaned['ride_min'] = divvy_cleaned['ride_length'].astype('timedelta64[m]')
```

```python
divvy_cleaned.head(3)
```

```python
divvy_cleaned.sort_values(['ride_length'],ascending=1).head(3)
```

```python
divvy_cleaned.loc[:, 'day_of_week'] = divvy_cleaned['started_at'].dt.dayofweek
divvy_cleaned.loc[:, 'Month'] = divvy_cleaned['started_at'].dt.month
divvy_cleaned.loc[:, 'Year'] = divvy_cleaned['started_at'].dt.year
```

```python
divvy_cleaned.head(3)
```

```python
divvy_cleaned['new_year_month'] = divvy_cleaned['started_at'].dt.strftime('%Y-%m')
```

```python
divvy_cleaned['Year_Month'] = divvy_cleaned['Year'].astype('str') + divvy_cleaned['Month'].astype('str')
```

```python
divvy_cleaned['Year_Month'] = pd.to_numeric(divvy_cleaned['Year_Month'])
```
